## COVID data Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def make_local(file):
    return file.replace('https://github.com/CSSEGISandData/COVID-19/blob/master/', '../COVID-19/')
```

```python
url = 'https://github.com/CSSEGISandData/COVID-19/blob/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
# file = '../COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv' 
confirmed_df = pd.read_csv(make_local(url))
print(confirmed_df.info())
confirmed_df.head()
```

```python
url = 'https://github.com/CSSEGISandData/COVID-19/blob/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'
deaths_df = pd.read_csv(make_local(url))
print(deaths_df.info())
deaths_df.head()
```
